```python
import wbgapi as wb
import pandas as pd
import matplotlib.pyplot as plt 
from labellines import labelLine, labelLines
```

# Find Codes

```python
# Get country codes

wb.economy.info()
```

```python
# Get indicator codes

wb.series.info()
```

```python
# Get trade percent of GDP (World)

trade_per_WLD = wb.data.DataFrame('NE.TRD.GNFS.ZS','WLD',mrv=20)
```

```python
# Plot trade percet of GDP (World)

fig, ax = plt.subplots(figsize=(20,10))

trade_per_WLD.T.plot(ax=ax)

ax.legend()

labelLines(ax.get_lines(), zorder=2.5)

plt.savefig('trade % GDP World.png')
```

Trade % of GDP of the world seems to be heading down. I am still suspicious since this is an aggregate.

```python
# Get trade % of GDP of largest western economies

trade_per_west = wb.data.DataFrame('NE.TRD.GNFS.ZS',['USA','UK','CAN','ITA','JAP','FRA','DEU'],mrv=20)
```

```python
fig, ax = plt.subplots(figsize=(20,10))

trade_per_west.T.plot(ax=ax)

ax.legend()

labelLines(ax.get_lines(), zorder=2.5)

plt.savefig('trade % GDP Western.png')
```

It seems suspicions were correct. Trade is going up in western countries with Germany the lead.

```python
# Get trade % of GDP for largest eastern economies

trade_per_east = wb.data.DataFrame('NE.TRD.GNFS.ZS',['CHN','IND',],mrv=20)
```

```python
fig, ax = plt.subplots(figsize=(20,10))

trade_per_east.T.plot(ax=ax)

ax.legend()

labelLines(ax.get_lines(), zorder=2.5)

plt.savefig('trade % GDP Eastern.png')
```

Interesting, seems decline is from the Asian countries and it has been happening for sometime even before 2008.

```python
# Get trade % of GDP East and West

trade_per_west_east = wb.data.DataFrame(
    'NE.TRD.GNFS.ZS',
    ['USA','UK','CAN','ITA','JAP','FRA','DEU','CHN','IND','WLD'],
    mrv=20)
```

```python
fig, ax = plt.subplots(figsize=(20,10))

trade_per_west_east.T.plot(ax=ax)

ax.legend()

labelLines(ax.get_lines(), zorder=2.5)
plt.title('trade % GDP Western + Eastern')
plt.savefig('trade % GDP Western + Eastern.png')
```
